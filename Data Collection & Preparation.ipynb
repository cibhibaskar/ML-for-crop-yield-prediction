## Climate Data

```python
df_rain = pd.read_csv(path+'/rainfall.csv')
df_rain.head()
```

```python
# check data types 
df_rain.info()
```

```python
df_rain['average_rain_fall_mm_per_year'] = pd.to_numeric(df_rain['average_rain_fall_mm_per_year'],
                                                         errors = 'coerce')
df_rain.info()
```

```python
df_rain.dropna(axis=0, inplace=True)
df_rain.info()
```

```python
df_rain.rename(columns={' Area' : 'Area', 'average_rain_fall_mm_per_year':'average_rain_fall'}, inplace=True)
```

```python
# Check Duplicate rows
duplicate = df_rain[df_rain.duplicated()]
duplicate
```

## Merge Yield and Climate data

```python
df = df.merge(df_rain, on=['Area', 'Year'])
df
```

```python
df.describe()
```

## Temperature Data

```python
df_temp = pd.read_csv(path+'/temp.csv')
df_temp.head()
```

```python
df_temp.info()
```

```python
df_temp.dropna(axis=0, inplace=True)
df_temp.info()
```

```python
df_temp.rename(columns={'year':'Year', 'country':'Area'}, inplace=True)
df_temp.head()
```

```python
# Check Duplicate rows
duplicate = df_temp[df_temp.duplicated()]
duplicate
```

```python
df_temp.drop_duplicates(inplace=True)
df_temp.info()
```

## Merge Yield and Temperature 

```python
df = df.merge(df_temp, on=['Area', 'Year'])
df
```
